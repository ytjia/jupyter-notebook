```python
import sys
import json

#getting all the correct agents for each image
ref = json.load(open('/Users/didi/Downloads/dev.json'))

structure = {}
all_agents = ()
for (image,s) in ref.items():
    image_name = str(image.split(".")[0])
    if image_name not in structure:
        agents = set()
        for r in s["frames"]:
            if r.has_key('agent') and r['agent'] != '':
                agents.add(r['agent'].encode("utf-8"))
        structure[image_name] = agents
structure            
```

```python
import re
file_path = 'dev.output'
f = open(file_path)
pred_results  = {}
pre_image = " "
image = ""
image_verb = ""
verb = ""
flag = 0
for line in f:
    words = re.split(r'\s+',line)
    image = words[0]
    image_verb = words[0].split(".")[0]
    verb = words[1]
    if image != pre_image:
        if "agent" in words:
            flag = 1
            agent = words[words.index("agent")+1]
            pred_results[image_verb] = [(verb,agent)]
        else:
            flag = 0
            pred_results[image_verb] = [(verb,'null')]
        pre_image = image
    else:
        if "agent" in words:
            flag = 1
            agent = words[words.index("agent")+1]
            pred_results[image_verb].append((verb,agent))
        else:
            flag = 0
            pred_results[image_verb].append((verb,'null'))
            
pred_results
```

```python
all_images = structure.keys()
all_images
```

```python
word_map = {}
with open('words.txt') as words_f:
    for line in words_f:
        words = line.strip().split('\t')
        word_map[words[0]] = words[1]
words_f.close()
word_map
```

```python
top_k = 25
correct_agent_count = 0
m_flag = -1
f_flag = -1
man_as_agent = 0
woman_as_agent = 0
man_wrongly_as_woman = 0 #supposed to be woman, but predicted as man
woman_wrongly_as_man = 0 #supposed to be man, but predicted as woman
wrong_verb_correct_gender = 0
wrong_verb_correct_agent = 0
correct_verb = 0
correct_gender = 0
M = ['man', 'male', 'boy']
F = ['woman', 'female', 'girl']
image_results = {}
all_imag = all_images[:]
for image in all_imag:
    correct_agent_count = 0
    m_flag = -1
    f_flag = -1
    man_as_agent = 0
    woman_as_agent = 0
    man_wrongly_as_woman = 0 
    woman_wrongly_as_man = 0
    wrong_verb_correct_gender = 0
    wrong_verb_correct_agent = 0
    correct_verb = 0
    correct_gender = 0
    true_agents = structure[image]
    true_gender = set()
    has_correct_agent = 0
#     print image
#     print true_agents
    true_agent_words = set()
    for ta in true_agents:
        true_agent_str = word_map[ta]
        true_agent_ws = re.split(' |, ',true_agent_str)
        for item in true_agent_ws:
            true_agent_words.add(item )
#     print true_agent_words
    
    for tw in true_agent_words: #true agent is man or woman; can be both
        if tw in M:
            m_flag = 1
            true_gender.add('m')
        elif tw in F:
            f_flag = 1
            true_gender.add('f')
    predict_agents = []
    predict_verbs = []
    img_verb = image.split("_")[0]
#     print true_gender
    for i in range(top_k):
        predict_agents.append(pred_results[image][i][1]) #predict_agents
        predict_verbs.append(pred_results[image][i][0])
#     print predict_agents
    for pv in predict_verbs:
        if pv == img_verb:
            correct_verb += 1
    for preagent in predict_agents:
        if preagent != 'null':
            pre_agent_str = word_map[preagent]
        pre_agent_words = re.split(' |, ',pre_agent_str) #agents written in words
#         print "predict_agents" 
#         print pre_agent_words
        for m in pre_agent_words:
            if m in M:
                man_as_agent += 1
                break
            elif m in F:
                woman_as_agent += 1
                break
        if preagent in true_agents:
            correct_agent_count += 1
            has_correct_agent = 1
            if predict_verbs[predict_agents.index(preagent)] != img_verb:
                wrong_verb_correct_agent += 1
        
        if len(true_gender) == 1 and 'm' in true_gender and woman_as_agent > 0:
            woman_wrongly_as_man = woman_as_agent
        elif len(true_gender) == 1 and 'f' in true_gender and man_as_agent > 0 :
            man_wrongly_as_woman = man_as_agent
            
        if 'm' in true_gender and man_as_agent > 0 or 'f' in true_gender and woman_as_agent > 0:
            correct_gender = man_as_agent + woman_as_agent
            if preagent in true_agents and predict_verbs[predict_agents.index(preagent)] != img_verb:
                wrong_verb_correct_gender += 1
#         elif true_gender == set():
#             correct_gender = -1
#             wrong_verb_correct_gender = -1
#         else:
#             correct_gender = 0
#             wrong_verb_correct_gender = 99999
            
#     if has_correct_agent == 0:
#         wrong_verb_correct_agent = 99999
        
    image_results[image] = (correct_agent_count, (top_k-correct_agent_count), man_as_agent, woman_as_agent, 
                            man_wrongly_as_woman, woman_wrongly_as_man, correct_verb, (top_k - correct_verb), 
                            correct_gender, wrong_verb_correct_gender, wrong_verb_correct_agent )
image_results           
```

```python
from operator import add
import copy
pre_img_verb = ''
all_results = {}
count_result = ((0,))*len(image_results.values()[0])
for img in sorted(image_results.keys()):
#     print "img: "+str(img)+"\t"+"pre_verb: "+pre_img_verb
    img_verb = img.split("_")[0]
    imr = image_results[img]
#     print imr
    if img_verb == pre_img_verb:
        count_result = copy.deepcopy(map(add, count_result, imr))
#         print 'xxxxxx'
#         print count_result
#         print '*****'
    else:
        if pre_img_verb != '':
            all_results[pre_img_verb] = copy.deepcopy(count_result)
#             print "%%%%%% "+pre_img_verb
#             print all_results[pre_img_verb]
            count_result = copy.deepcopy(imr)
        else:
            count_result = copy.deepcopy(map(add, count_result, imr))
#             print '&&&&&'
#             print count_result
        
        pre_img_verb = img_verb
all_results[pre_img_verb] = count_result
```
